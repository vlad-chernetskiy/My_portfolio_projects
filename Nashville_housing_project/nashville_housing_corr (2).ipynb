```python
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.DataFrame(pd.read_csv("C:/Users/borsh/OneDrive/Рабочий стол/portfolio project/sales.csv"))
df.head()
```

```python
plt.scatter(x=df['acreage'], y=df['saleprice'])
plt.title('Acreage/SalePrice')
plt.xlabel('Acreage')
plt.ylabel('SalePrice')
```

```python
plt.scatter(x=df['totalvalue'], y=df['saleprice'])
plt.title('TotalValue/SalePrice')
plt.xlabel('TotalValue')
plt.ylabel('SalePrice')
```

Our hypotheses is a saleprice has correlation with acreage and Total Value, moreover between total value it look stronger correlation 

```python
sb.regplot(x=df['acreage'], y=df['saleprice'],data=df,line_kws={'color':'red'})
plt.title('Acreage/SalePrice')
plt.xlabel('Acreage')
plt.ylabel('SalePrice')
```

```python
sb.regplot(x=df['totalvalue'], y=df['saleprice'],data=df,line_kws={'color':'red'})
plt.title('TotalValue/SalePrice')
plt.xlabel('TotalValue')
plt.ylabel('SalePrice')
```

Now lets build correlation matrix and check our hypothesis

```python
corr_matrix=df.corr(numeric_only=True)
sb.heatmap(corr_matrix,annot=True)
plt.show()
```

We found that total value has the highest correlation with saleprice that every metric, but acreage has no impact to sale price. That means that our hypothesis was particualry wrong.
